```python
import folium
import pandas as pd
import numpy as np
```

```python
LON = pd.Series(np.random.randint(1,20,size=1000))
LAT = pd.Series(np.random.randint(56,60,size=1000))
LON.name = 'LON'
LAT.name = 'LAT'

df= pd.concat([LON,LAT], axis=1)
df
```

```python
map=folium.Map(location=[df['LAT'].mean(),df['LON'].mean()],zoom_start=6,tiles='Mapbox bright')

fg=folium.FeatureGroup(name="Volcano Locations")
for lat,lon in zip(df['LAT'],df['LON']):
    fg.add_child(folium.Marker(location=[lat,lon],icon=folium.Icon(color='red')))

mg=folium.FeatureGroup(name="Volcano Locations")
for lat,lon in zip(df['LAT']+2,df['LON']+2):
    mg.add_child(folium.Marker(location=[lat,lon],icon=folium.Icon(color='green')))    
    
map.add_child(fg)
map.add_child(mg)

map.save(outfile='map.html')
```
